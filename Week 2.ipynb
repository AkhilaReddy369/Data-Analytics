```python
import pandas as pd

# Load the CSV file into a DataFrame, setting the 'INSTNM' column as the index
# 'INSTNM' represents the institution name, so it's a good candidate for the index
college = pd.read_csv('MERGED2008_09_PP.csv', index_col='UNITID')
```

```python
print("# Display the first five rows of the dataset")
print(college.head())
```

```python
num_rows, num_columns = college.shape
print(f"The dataset contains {num_rows} rows and {num_columns} columns.")
```

```python
print(college.columns.tolist())
```

```python
print(college.dtypes)
```

```python
missing_values = college.isnull().sum()
print(missing_values[missing_values > 0])
```

```python
print(college.describe())
```

```python
state_distribution = college['STABBR'].value_counts()
print(state_distribution)
```

```python
unique_institutions = college.index.nunique()
print(f"There are {unique_institutions} unique institutions in the dataset.")
```

```python
duplicates = college.duplicated().sum()
print(f"There are {duplicates} duplicate rows in the dataset.")
```

```python
if 'Year' in college.columns:
    year_range = college['Year'].min(), college['Year'].max()
    print(f"The range of years in the dataset is from {year_range[0]} to {year_range[1]}.")
else:
    print("No 'Year' column found in the dataset.")
```

```python
selected_columns =['OPEID','INSTNM','CITY','STABBR','ZIP','REGION','NUMBRANCH','HIGHDEG','LATITUDE','LONGITUDE','CCBASIC','CCSIZSET','HBCU','MENONLY','WOMENONLY','RELAFFIL','ADM_RATE','SAT_AVG','DISTANCEONLY','UG','UGDS','COSTT4_A', 'COSTT4_P','TUITIONFEE_IN','TUITIONFEE_OUT','AVGFACSAL']
```

```python
new_df = college[selected_columns]
print(new_df.head())
```

```python
# Import the 'display' function from the IPython.display module
from IPython.display import display

# Set Pandas options to not truncate columns
pd.set_option('display.max_columns', None)  # Show all columns
pd.set_option('display.expand_frame_repr', False)  # Disable wrapping of DataFrame to multiple lines

# Use the 'display' function to visually display the DataFrame 'df'
display(new_df)
```
